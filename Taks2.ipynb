```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
```

```python
def update_grid(grid):
    rows, cols = grid.shape
    new_grid = np.zeros((rows, cols))

    for y in range(rows):
        for x in range(cols):
            live_neighbors = 0
            
            for i in range(-1, 2):
                for j in range(-1, 2):
                    if i == 0 and j == 0:
                        continue 
                    ni, nj = y + i, x + j
                    if 0 <= ni < rows and 0 <= nj < cols:
                        live_neighbors += grid[ni, nj]


            if grid[y, x] == 1:
                if live_neighbors == 2 or live_neighbors == 3:
                    new_grid[y, x] = 1
            else:
                if live_neighbors == 3:
                    new_grid[y, x] = 1
    return new_grid

grid_size = 20

current_grid = np.random.choice([0, 1], size=(grid_size, grid_size))

for _ in range(20):
    current_grid = update_grid(current_grid)

fig, ax = plt.subplots(figsize=(8, 8))

cmap = ListedColormap(['#d3d3d3', '#397a2b'])

ax.pcolormesh(current_grid, cmap=cmap, edgecolors='white', linewidth=2)

ax.set_aspect('equal') 
ax.invert_yaxis()      
ax.axis('off')         

plt.show()
```
