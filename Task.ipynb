```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
Chisla = [random.randint(0, 20) for _ in range(20)]  
print(Chisla)
```

```python
n = len(Chisla)
left_max = [0] * n
right_max = [0] * n

left_max[0] = Chisla[0]
for i in range(1, n):
    left_max[i] = max(left_max[i-1], Chisla[i])
right_max[n-1] = Chisla[n-1]
for i in range(n-2, -1, -1):
    right_max[i] = max(right_max[i+1], Chisla[i])

water_levels = [min(left_max[i], right_max[i]) for i in range(n)]
depths = [max(0, water_levels[i] - Chisla[i]) for i in range(n)]

max_d = max(depths)
idx = depths.index(max_d)
top_line = water_levels[idx]

start, end = idx, idx
while start > 0 and Chisla[start] < top_line: start -= 1
while end < n - 1 and Chisla[end] < top_line: end += 1

river_idx = list(range(start, end + 1))
river_val = [Chisla[i] for i in river_idx]
bottom_line = min(river_val)
```

```python
plt.plot(Chisla, ) 
plt.plot(river_idx, river_val, color='red', linewidth=2.5)
plt.axhline(y=top_line, color='green', linestyle='--') 
plt.axhline(y=bottom_line, color='green', linestyle='--')
plt.show()
```
